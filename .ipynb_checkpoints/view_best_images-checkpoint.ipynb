```python
import matplotlib.pyplot as plt
```

```python
score_path = 'Results/fft_metric_output.txt'
```

```python
im_dict = {}
with open(score_path,'r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.replace('\n','')
        image,score = line.split(',')
        im_dict[image] = float(score)
```

## Show Sharpest Images

```python
sharpest_x_images = 20
blurry_images = list(im_dict.keys())[-sharpest_x_images:]

for file in blurry_images:
    im = plt.imread(file)
    print(file)
    plt.imshow(im,cmap='gray')
    plt.show()
```

## Show Blurriest Images

```python
blurriest_x_images = 20
blurry_images = list(im_dict.keys())[:blurriest_x_images]
for file in blurry_images:
    #if im_dict[file] < 1000:
        #display(Image(filename=file,width=130, height=130))  
    im = plt.imread(file)
    print(file)
    plt.imshow(im,cmap='gray')
    plt.show()
```
